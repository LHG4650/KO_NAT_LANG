```python
import pandas as pd
import io
import matplotlib.pyplot as plt

# csv_text를 DataFrame으로 변환
df = pd.read_csv(io.StringIO(csv_text))

# poison_ratio를 문자열로 변환해서 카테고리 취급
df['poison_ratio_str'] = df['poison_ratio'].astype(str)

# sampling_ratio별 그룹
sampling_ratios = df['sampling_ratio'].unique()

# y축 최소값 설정 (원하는 값으로 수정 가능)
y_min = 0.8

# 분석할 컬럼 리스트 (pixel_auroc, image_auroc 둘 다)
metrics = ['image_auroc', 'pixel_auroc']

for metric in metrics:
    print(f"==== {metric} ====")

    for sampling_ratio in sorted(sampling_ratios):
        sub_df = df[df['sampling_ratio'] == sampling_ratio]

        fig, ax = plt.subplots(figsize=(10,6))

        # poison_ratio별로 해당 metric 리스트 만들기
        poison_groups = sub_df.groupby('poison_ratio_str')[metric].apply(list)

        # 박스플롯 그리기
        positions = range(len(poison_groups))  # x축 위치를 0,1,2,3,...로 고정
        ax.boxplot(poison_groups, positions=positions)

        # 평균값 계산
        means = sub_df.groupby('poison_ratio_str')[metric].mean()

        # 추세선 그리기 (positions 기준으로)
        ax.plot(positions, means.values, marker='o', linestyle='-', label=f'Sampling Ratio {sampling_ratio}')

        # x축 설정
        ax.set_xticks(positions)
        ax.set_xticklabels(poison_groups.index)

        # 제목, 축 이름
        ax.set_title(f'{metric.upper()} - Sampling Ratio = {sampling_ratio}')
        ax.set_xlabel('Poison Ratio')
        ax.set_ylabel(metric.upper())

        # y축 범위
        ax.set_ylim(y_min, 1.0)

        ax.grid(True)
        ax.legend()
        plt.show()
```
